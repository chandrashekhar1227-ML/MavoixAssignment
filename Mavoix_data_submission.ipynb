```python
import numpy as np
import pandas as pd
train=pd.read_excel('C:/Users/chandu prakash/Downloads/mavoix_ml_sample_dataset.xlsx')
```

```python
#Feature engineering (Generating features from Other skills column)
oskills = train["Other skills"]
cols = [' Java', ' C++ Programming',' C Programming', ' MS-Excel', ' Data Structures',' Data Analytics',
        ' Database Management System (DBMS)',
        ' Natural Language Processing (NLP)']
```

```python
dit = {}
for j in cols:
    l = [(j in i if type(i) == str else False) for i in oskills]
    dit[j] = l
```

```python
new_df = pd.DataFrame(dit, columns=cols)
```

```python
new_df.to_excel("mavoix_feature_engineering.xlsx")
```

```python
new_train=train.iloc[:,[2,3,4,5,6,7,8,9,11,12,13,14,15]]
```

```python
final_df=pd.concat([new_train,df1], axis=1)
```

```python
final_df.isnull().sum()
```

```python
from sklearn.cluster import KMeans
```

```python
km = KMeans(n_clusters=2)
y_predicted = km.fit_predict(final_df)
y_predicted
```

```python
np.unique(y_predicted,return_counts=True)
```

```python
# Elbow method for determining optimal value K
sse = []
k_rng = range(1,10)
for k in k_rng:
    km = KMeans(n_clusters=k)
    km.fit(final_df)
    sse.append(km.inertia_)
```

```python
from matplotlib import pyplot as plt
%matplotlib inline
```

```python
plt.xlabel('K')
plt.ylabel('Sum of squared error')
plt.plot(k_rng,sse)
```

```python
#Optimal value is equal to 2(from the above graph)[K=2]
```
